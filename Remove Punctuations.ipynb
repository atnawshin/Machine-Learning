```python
import pandas as pd
import nltk
import re
from nltk.corpus import stopwords

df = pd.read_excel('sample.xlsx')

df.head()
```

```python
#remove stop words

";".join(stopwords.words('english'))
stop_words=set(stopwords.words('english'))

def remove_stop(x):
    return ",".join([word for word in str(x).split() if word not in stop_words])
df['Post']=df['Post'].apply(lambda x : remove_stop(x))
```

```python
df.head()
```

```python
#Data cleaning function
#removing unnecessary punctuation
def process_posts(Post): 
    Post = re.sub('[^\u0980-\u09FF]',' ',str(Post)) 
    return Post
```

```python
# Apply the function into the dataframe
df['cleaned'] = df['Post'].apply(process_posts)  

# print some cleaned reviews from the dataset
sample_df = [10,11,25,8,9,6]
for i in sample_df:
     print('Original:\n',df.Post[i],'\nCleaned:\n',
           df.cleaned[i],'\n','Sentiment:-- ',df.Label[i],'\n')
```
