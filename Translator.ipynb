```python
import pandas as pd
#Reading the excel file
df = pd.read_excel('Restaurant.xlsx')
df.head()   # we are going to translate the column last_comment_from_farmer
```

```python
#Translate the other languages to english
import googletrans
from googletrans import *
translator = googletrans.Translator()

df['Text'] = df['Text'].astype(str) #changing datatype to string
df['translated_Text'] = df['Text'].apply(translator.translate, src='auto', dest='en').apply(getattr, args=('text',))
df
```
